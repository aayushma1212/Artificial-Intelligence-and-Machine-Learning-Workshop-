Task1: Create a Python program that converts between different units of measurement

```python
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    Parameters:
    conversion_type (int): The option selected by the user for conversion.
    value (float): The number to convert.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:
        return value * 3.28084  # meters to feet
    elif conversion_type == 2:
        return value / 3.28084  # feet to meters
    elif conversion_type == 3:
        return value * 2.20462  # kg to pounds
    elif conversion_type == 4:
        return value / 2.20462  # pounds to kg
    elif conversion_type == 5:
        return value * 0.264172  # liters to gallons
    elif conversion_type == 6:
        return value / 0.264172  # gallons to liters
    else:
        return None


try:
    print("Choose conversion type:")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    choice = int(input("Enter option number: "))
    value = float(input("Enter value to convert: "))

    result = convert_units(choice, value)

    if result is None:
        print("Invalid conversion type.")
    else:
        print("Converted value:", result)

except ValueError:
    print("Error: Please enter a valid number.")
```

Task2: Create a Python program that performs various mathematical operations on a list of numbers.

```python
def find_sum(numbers):
    """Calculates the sum of a list of numbers."""
    return sum(numbers)

def find_average(numbers):
    """Calculates the average of a list of numbers."""
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    """Finds the maximum value in a list of numbers."""
    return max(numbers)

def find_minimum(numbers):
    """Finds the minimum value in a list of numbers."""
    return min(numbers)

def math_operations_program():
    print("1. Sum\n2. Average\n3. Maximum\n4. Minimum")

    try:
        choice = int(input("Select an operation (1-4): "))

        raw_input = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw_input.split()]

        if not numbers:
            print("Error: The list cannot be empty.")
            return

        if choice == 1:
            print(f"Result (Sum): {find_sum(numbers)}")
        elif choice == 2:
            print(f"Result (Average): {find_average(numbers)}")
        elif choice == 3:
            print(f"Result (Maximum): {find_maximum(numbers)}")
        elif choice == 4:
            print(f"Result (Minimum): {find_minimum(numbers)}")
        else:
            print("Invalid selection. Please choose 1, 2, 3, or 4.")

    except ValueError:
        print("Error: Please enter only numeric values (numbers).")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

math_operations_program()
```

**NESTED LIST**

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension

```python
def flatten(lst):
    flat_list = []

    for sublist in lst:
        for item in sublist:
            flat_list.append(item)

    return flat_list

data = [[1, 2], [3, 4], [5]]
print(flatten(data))
```

Write a Python function that extracts a specific element from a nested list given its indices

```python
def access_nested_element(lst, indices):
    return lst[indices[0]][indices[1]]
lst = [[1,2,3],[4,5,6],[7,8,9]]
indices = [1,2]

print(access_nested_element(lst, indices))
```

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth)

```python
def sum_nested(lst):
    total = 0

    for item in lst:
        if isinstance(item, list):
            total = total + sum_nested(item)
        else:
            total = total + item

    return total

data = [[1,2],[3,[4,5]],6]
print(sum_nested(data))
```

Write a Python function that removes all occurrences of a specific element from a nested list

```python
def remove_element(lst, elem):
    new_list = []

    for sublist in lst:
        temp = []
        for item in sublist:
            if item != elem:
                temp.append(item)
        new_list.append(temp)

    return new_list

lst = [[1,2],[3,2],[4,5]]
print(remove_element(lst,2))
```

Write a Python function that finds the maximum element in a nested list (regardless of depth)

```python
def find_max(lst):
    maximum = float("-inf")

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)
            if value > maximum:
                maximum = value
        else:
            if item > maximum:
                maximum = item

    return maximum

data = [[1,2],[3,[4,5]],6]
print(find_max(data))
```

Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    count = 0

    for item in lst:
        if isinstance(item, list):
            count = count + count_occurrences(item, elem)
        else:
            if item == elem:
                count = count + 1

    return count
lst = [[1,2],[2,3],[2,4]]
print(count_occurrences(lst,2))
```

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth

```python
def deep_flatten(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result = result + deep_flatten(item)
        else:
            result.append(item)

    return result

data = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(deep_flatten(data))
```

Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):

    def helper(lst):
        total = 0
        count = 0

        for item in lst:
            if isinstance(item, list):
                t, c = helper(item)
                total = total + t
                count = count + c
            else:
                total = total + item
                count = count + 1

        return total, count

    total, count = helper(lst)
    return total / count

data = [[1,2],[3,4],[5,6]]
print(average_nested(data))
```
